```python
# Deliverable Five : Compare the Data 

#a) display the average budget for each school type 

school_type_df = student_df.groupby("school_type")

school_type_df[["school_budget"]].mean()
```

```python
#b) find the total number of students at each school and sort those values from largest to smallest 

student_df["student_count"] = student_df.groupby("school_name")["school_name"].transform('count')

student_df.loc[: ,[ "school_name", "student_count"]]



#first I made the new variable student_count using the groupby function and the transform function to add the 
#colum for this new variable, then I looked to see if it worked by using .loc


```

```python
#b) continued

#here I made a new dataframe, students_df, and set it equal to a count of the student_df sorted 
#by student_count and grouped by school name

students_df = student_df.sort_values("student_count").groupby("school_name").count()

students_df.sort_values(by= "student_count", ascending=False).loc[:, ["student_count"]]

#the second line of code sorts the students_df by student count and makes it go 
#in descending order, and then i use .loc to find the student_count 
#i think i might have done too much here but it worked #so 
```

```python
#c) find the average math score by grade for each school type 

#first i make the average math score equal to the student_df grouped by school type and grade 
#and the aggregate demand is the mean. 
#then I realized after running it the first time 
#that I needed to use the round function to round to the nearest whole number
#so that it would look like the table we saw. 
average_math_score = student_df.groupby(["school_type", "grade"]).mean().round(0) #round(0) is to the whole number 
#round(1) is to the tenths

average_math_score.loc[:, ["math_score"]]
```
